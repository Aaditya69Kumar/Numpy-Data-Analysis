# With NumPy:-

```python
import numpy as np

math_grades = np.array([85, 92, 78, 88])
science_grades = np.array([88, 94, 82, 90])

total = math_grades + science_grades
print(total)
```

# Creating Arrays:- four ways

```python
import numpy as np

# Method 1: From list
grades = np.array([85, 92, 78, 88, 95])
print("Grades:", grades)

# Method 2: Zeros (used for ML initialization)
zeros_array = np.zeros(5)  # [0., 0., 0., 0., 0.]
print("Zeros:", zeros_array)

# Method 3: Ones
ones_array = np.ones(5)
print("Ones:", ones_array)

# Method 4: Range
range_array = np.arange(0, 10, 2)  # [0, 2, 4, 6, 8]
print("Range:", range_array)
```

# Now Let's work with some real data

```python
# Create arrays for our class
student_ids = np.arange(1, 11) # what is the output of this ???
math_scores = np.array([85, 92, 78, 88, 95, 82, 76, 91, 84, 89])
science_scores = np.array([88, 94, 82, 90, 93, 85, 79, 95, 87, 92])

print("Student IDs:", student_ids)
print("Math Scores:", math_scores)
print("Science Scores:", science_scores)
```

# Aggregation (Summary statistics)

```python
print("\n--- Class Statistics ---")
print("Math average:", np.mean(math_scores))
print("Math highest:", np.max(math_scores))
print("Math lowest:", np.min(math_scores))
print("Math standard deviation:", np.std(math_scores))
print("Sum of all math scores:", np.sum(math_scores))
```

# Indexing with Conditions

```python
# Find all students with math > 90
top_math_students = student_ids[math_scores > 90]
print("Top math students (ID):", top_math_students)

# Find scores of students who need help
low_scores = math_scores[math_scores < 80]
print("Low scores that need help:", low_scores)

# Multiple conditions
good_students = student_ids[(math_scores > 85) & (science_scores > 90)]
print("Good in both subjects:", good_students)
```

# Reshaping

```python
# Our scores as a flat list
all_scores = np.array([85, 92, 78, 88, 95, 82, 76, 91, 84, 89,
                       88, 94, 82, 90, 93, 85, 79, 95, 87, 92])

# Reshape to 2D: 10 students × 2 subjects
score_matrix = all_scores.reshape(10, 2)
print("\nScore Matrix (Students × Subjects):")
print(score_matrix)
print("Shape:", score_matrix.shape)  # (10, 2)
```

# Feature Matrices (x) ->

```python
# In ML, data must be 2D arrays
# Each row = one student
# Each column = one feature (score, age, etc.)

# Create feature matrix for ML
X = np.array([
    [85, 20],  # Student 1: math=85, age=20
    [92, 21],  # Student 2: math=92, age=21
    [78, 22],  # Student 3: math=78, age=22
    [88, 20],  # Student 4: math=88, age=20
])

print("Feature matrix X:")
print(X)
print("Shape (rows, cols):", X.shape)
```
