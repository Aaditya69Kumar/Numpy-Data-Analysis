# Task:- NumPy Functions Exploration

```python
import numpy as np


np.random.seed(42)


num_students = 200


student_ids = np.arange(1, num_students + 1)
math_scores = np.random.normal(loc=75, scale=15, size=num_students).astype(int)
science_scores = np.random.normal(loc=78, scale=12, size=num_students).astype(int)
english_scores = np.random.normal(loc=72, scale=10, size=num_students).astype(int)
attendance = np.random.randint(60, 101, size=num_students)
study_hours = np.random.uniform(1, 20, size=num_students).round(1)
has_tutor = np.random.choice([0, 1], size=num_students, p=[0.7, 0.3])  # 0=No, 1=Yes


math_scores = np.clip(math_scores, 0, 100)
science_scores = np.clip(science_scores, 0, 100)
english_scores = np.clip(english_scores, 0, 100)

print(" Dataset created successfully!")
print(f"Total students: {num_students}")
```

# TASK 2: Statistical Analysis


```python
# Task: Calculate these statistics for math scores:
# 1. Mean score
# 2. Standard deviation
# 3. Maximum score
# 4. Minimum score
# 5. Median score
# Use: np.mean(), np.std(), np.max(), np.min(), np.median()

# Your code here:
math_mean = np.mean(math_scores)
math_std = np.std(math_scores)
math_max = np.max(math_scores)
math_min = np.min(math_scores)
math_median = np.median(math_scores)

print(f"Math Scores - Mean: {math_mean:.2f}, Std: {math_std:.2f}, Range: {math_min}-{math_max}, Median: {math_median:.2f}")
```

```python
# Task: Find:
# 1. Students in top 10% (90th percentile)
# 2. Students in bottom 25% (25th percentile)
# Use: np.percentile()

# Your code here:
# Calculate total_scores, as it was not defined in a preceding cell.
total_scores = math_scores + science_scores + english_scores

top_10_threshold = np.percentile(total_scores, 90)
bottom_25_threshold = np.percentile(total_scores, 25)

top_students = student_ids[total_scores >= top_10_threshold]
bottom_students = student_ids[total_scores <= bottom_25_threshold]

print(f"Top 10% students (IDs, score >= {top_10_threshold:.2f}): {top_students}")
print(f"Bottom 25% students (IDs, score <= {bottom_25_threshold:.2f}): {bottom_students}")
```
